```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pandas_profiling import ProfileReport
```

```python
data_df = df = pd.read_csv('data.csv', delimiter='\t', quotechar='"')
```

```python
data_df
```

```python
data_df.info()
```

```python
data_df['QUADRO'].unique()
```

```python
plt.figure(figsize=(10, 10))  # Adjust the figsize as per your preference
ax = sns.countplot(data=data_df, y='AFLUENCIA')

# Add labels on top of bars
for p in ax.patches:
    ax.annotate(format(p.get_width(), '.0f'), (p.get_x() + p.get_width(), p.get_y() + p.get_height() / 2.), ha='left', va='center')

plt.show()
```
